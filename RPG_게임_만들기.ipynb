```python
from random import *

class Character:
  def __init__(self, name, lv, hp, ap, dp):
    self.name = name
    self.lv = lv
    self.hp = hp
    self.ap = ap
    self.dp = dp 

  def is_alive(self):
    return self.hp > 0
      
  def attack_target(self, target):
    damage = randint(1, self.ap)
    print(f"{self.name}(ì´)ê°€ {target.name}ì—ê²Œ ì…íŒ ë°ë¯¸ì§€: {damage}")
    target.take_damage(damage)

  def take_damage(self, damage):
    if self.dp > damage:
      print(f"{self.name}ì˜ ë°©ì–´ë ¥ì´ ë†’ì•„ í”¼í•´ë¥¼ ì…ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤.")
      return
    else:
      self.hp -= damage
      self.hp = max(0, self.hp)
      print(f"{self.name}(ì´)ê°€ ë°›ì€ ë°ë¯¸ì§€: {damage} | ë‚¨ì€ ì²´ë ¥: {self.hp}\n")
  
  def status(self): #ìƒíƒœì°½
    print('=' * 55)
    print(f" {self.name} | ë ˆë²¨: {self.lv} | ì²´ë ¥: {self.hp} | ê³µê²©ë ¥: {self.ap} | ë°©ì–´ë ¥: {self.dp}")
    print('=' * 55)


class Player(Character):
  def __init__(self, name):
    super().__init__(name, 1, 100, 25, 5)
    self.exp = 0

  def gain_experience(self, exp):
    self.exp += exp
    print(f"íšë“ ê²½í—˜ì¹˜:{exp} | í˜„ì¬ ê²½í—˜ì¹˜:{self.exp}")
    if self.exp >= 50:
      self.level_up()

  def level_up(self):
    if self.exp >= 50:
      self.exp -= 50 #ì´ˆê³¼ë¶„
      self.lv += 1
      self.ap += 10
      self.dp += 5
      self.hp = 100 #ë ˆë²¨ì—… ì‹œ í’€ì¶©ì „
      print(f"ë ˆë²¨ì´ ì˜¬ëìŠµë‹ˆë‹¤!\n í˜„ì¬ ë ˆë²¨: {self.lv} | í˜„ì¬ ê²½í—˜ì¹˜: {self.exp} | ì²´ë ¥: {self.hp}")


class Monster(Character):
  def __init__(self, name, monster_dict):    
    lv = monster_dict.get(name)
    hp = randint(10, 30) * lv
    ap = randint(5,20) * lv
    dp = randint(1,5) * lv
    super().__init__(name, lv, hp, ap, dp)
    


def battle(player, monster):
  while player.is_alive() and monster.is_alive():
    monster.status()

    print(f"\nê³µê²© [1] | ìƒíƒœì°½ [0]\n")
    while True:
      try:
        press = int(input("ë²ˆí˜¸ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”."))
        if press in [0, 1]:
          break
        else:
          print("ì˜ëª»ëœ ê°’ì…ë‹ˆë‹¤. 1, 0 ì¤‘ í•˜ë‚˜ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”.")
      except ValueError:
        print("ìˆ«ìë§Œ ì…ë ¥ ê°€ëŠ¥í•©ë‹ˆë‹¤.")

    if press == 1:
      player.attack_target(monster)
      if monster.is_alive():
        monster.attack_target(player)
    elif press == 0:
      player.status()
      continue
    else:
      print("ì˜ëª»ëœ ê°’ì…ë‹ˆë‹¤. 1, 0 ì¤‘ í•˜ë‚˜ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”.")
      continue     

  if player.is_alive():
    print("ì „íˆ¬ ìŠ¹ë¦¬!")
    player.gain_experience(monster.lv * 20)
  else:
    print("ì „íˆ¬ íŒ¨ë°°...")

def main():
  monster_dict = {"ìŠ¬ë¼ì„": 1, "ê³ ë¸”ë¦°": 2, "ì˜¤í¬": 3, "íŠ¸ë¡¤": 4, "ìš©": 5}
  print("ëª¨í—˜ì„ ì‹œì‘í•©ë‹ˆë‹¤...")
  player_name = input("ìš©ì‚¬ì˜ ì´ë¦„ì€?: ")
  player = Player(player_name)


  while player.is_alive() and monster_dict: #ë¹„ì–´ìˆë‹¤ë©´ False
    monster_name = min(monster_dict, key = monster_dict.get)
    monster = Monster(monster_name, monster_dict)
    print(f"\n{monster.name}(ì´)ê°€ ë‚˜íƒ€ë‚¬ìŠµë‹ˆë‹¤!\n")
    battle(player, monster)
    
    del monster_dict[monster_name]

  if player.is_alive():
    print("ìš©ì‚¬ê°€ ëª¬ìŠ¤í„°ë¥¼ ëª¨ë‘ ë¬´ì°”ë €ìŠµë‹ˆë‹¤! \nì™•êµ­ì— í‰í™”ê°€ ì°¾ì•„ì™”ìŠµë‹ˆë‹¤.")
  else:
    print("GAME OVER")

main()
```
